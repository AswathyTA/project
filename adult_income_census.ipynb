```python
#outliers present in fnlwgt
#IQR=q3-q1
q1=1.178270e+05
q3=2.370510e+05
IQR=q3-q1
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
upper_array=np.where(d['fnlwgt']>=upper_limit)[0]
lower_array=np.where(d['fnlwgt']<=lower_limit)[0]
d.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
for i in d.columns:
  d[i]=lab.fit_transform(d[i])
d.head()
```

```python
#feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
x=d.drop(['income'],axis=1)
y=d['income']
chi_scores=chi2(x,y)
chi_scores
```

```python
chi_values = pd.Series(chi_scores[0], index=x.columns)
chi_values.sort_values(ascending=False, inplace=True)
chi_values.plot.bar()
plt.yticks(chi_scores[0])
```

```python
#higher the chi_square value higher the importance from the graph race,nativecountry and workclass are of least importance
d.drop(columns=['race','native.country','workclass'],axis=1,inplace=True)
d
```

```python
x=d.iloc[:,:-1].values
y=d.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
o1=StandardScaler()
o1.fit(x_train)
x_train=o1.transform(x_train)
x_test=o1.transform(x_test)
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
ypred=model.predict(x_test)
ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
mat=confusion_matrix(y_test,ypred)
mat
```

```python
label=['0','1']
dis=ConfusionMatrixDisplay(mat,display_labels=label)
dis.plot()
```

```python
score=accuracy_score(y_test,ypred)

score
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,criterion='gini')
model.fit(x_train,y_train)
ypred=model.predict(x_test)
ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,ypred)
mat
```

```python
score=accuracy_score(y_test,ypred)
score
```

```python
from sklearn.model_selection import GridSearchCV

params= {'n_estimators':[10,100],'criterion':['gini','entropy']}
gd=GridSearchCV(model,params,cv=10,scoring='accuracy')
gd.fit(x_train,y_train)
```
