###                         **  INDIVIDUAL CARBON FOOTPRINT PREDICTION**





CARBON FOOTPRINT is a measure of carbondioxide emissions released into the atmosphere as a result of a particular activity done by an individual,organisation or community.It is expressed as measure of weights.
The carbon footprint is a valuable tool for measuring the contribution to climate change by an individual, organisations, products and services, and more.


DATASET DETAILS

The data contains  information about the day to day activities of an individual and how it effects their carbon footprint.


FEATURE INFO:
1.'Body Type': Body type

2.'Sex': Gender.

3.'Diet': Diet.

4.'How Often Shower': Frequency of showering

5.'Heating Energy Source': Residential heating energy

6.'Transport': Transportation preference.

7.'Vehicle Type': Vehicle fuel type.

8.'Social Activity': Frequency of participating in social activities.

9.'Monthly Grocery Bill': Monthly amount spent on groceries, in dollars.

10.'Frequency of Traveling by Air': Frequency of using aircraft in the
                                    last month.

11.'Vehicle Monthly Distance Km': The kilometers traveled by vehicle in
                                  the last month.

12.'Waste Bag Size': Size of the garbage bag

13.'Waste Bag Weekly Count': The amount of garbage thrown away in the
                             last week.

14.'How Long TV PC Daily Hour': Daily time spent in front of TV or PC.

15.'How Many New Clothes Monthly': Number of clothes purchased monthly.

16.'How Long Internet Daily Hour': Time spent on the Internet daily.

17.'Energy efficiency': Whether or not you care about purchasing energy
                         efficient devices.

18.'Recycling': The wastes it recycles.

19.'Cooking_With': Devices used in cooking.

20.'CarbonEmission': Dependent variable, total carbon emissions

DATA PREPROCESSING

```python
d.drop_duplicates().shape
```

```python
#checking for missing values
d.isna().sum()
```

```python
#dropping the column Vehicle Type because of too much null values
d.drop(['Vehicle Type'],axis=1,inplace=True)
d.head()
```

```python
#in the features 'Recycling' and 'Cooking with' the values are list in a string format to read it as lists
d['Recycling']=d['Recycling'].apply(ast.literal_eval)
d['Cooking_With']=d['Cooking_With'].apply(ast.literal_eval)


#converting lists in to comma separated values
d['Recycling']=d['Recycling'].apply(lambda x:','.join(x))

d['Cooking_With']=d['Cooking_With'].apply(lambda x:','.join(x))


#creating dummy variables for this transformed columns
recycling_dummies=d['Recycling'].str.get_dummies(sep=',')
cooking_with_dummies=d['Cooking_With'].str.get_dummies(sep=',')


#concatinating the dummy variables with the original dataframe after dropping the object columns
d=pd.concat([d.drop(['Recycling','Cooking_With'],axis=1),recycling_dummies,cooking_with_dummies],axis=1)
d.head()

```

```python
#converting all object columns into lowercase for consistency
text_columns=d.select_dtypes(include=['object']).columns

d[text_columns]=d[text_columns].apply(lambda x:x.str.lower())


#converting all feature name into lower case
#converting all space into underscore
d.columns=d.columns.str.replace(' ','_').str.lower()
d.head()
```

```python
#Checking for outliers
sns.boxplot(data=d,)
plt.xticks(rotation=90)
```

```python
# outliers present in vehicle_monthly_distance_km and carbon emission
# removing the outliers using IQR method
# IQR=q3-q1==>(inter quartile range)

q1 = np.percentile(d['vehicle_monthly_distance_km'], 25, method='midpoint')
q3 = np.percentile(d['vehicle_monthly_distance_km'], 75, method='midpoint')
print(q1)
print(q3)
```

```python
IQR=q3-q1
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
d['vehicle_monthly_distance_km'] = d['vehicle_monthly_distance_km'].astype(float)
d=d[~((d['vehicle_monthly_distance_km']< (lower_limit)) |(d['vehicle_monthly_distance_km'] > (upper_limit)))]
d.shape
```

```python
d['carbonemission']=d['carbonemission'].astype(float)
Q1=np.percentile(d['carbonemission'],25,method='midpoint')
Q3=np.percentile(d['carbonemission'],75,method='midpoint')
iqr=Q3-Q1
upper=Q3+iqr*1.5
lower=Q1-iqr*1.5

d=d[~((d['carbonemission']<(lower)) |(d['carbonemission']>(upper)))]
d.shape
```

```python
#converting the object column to numerical column
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
for i in d.columns:
  d[i]=lab.fit_transform(d[i])
d.tail()
```

Feature selection using chi-square test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
x=d.drop(['carbonemission'],axis=1)
y=d['carbonemission']
chi_scores=chi2(x,y)
chi_scores
```

```python
chi2_selector = SelectKBest(chi2, k=20)
X_kbest = chi2_selector.fit_transform(x, y)
selected_features = chi2_selector.get_support(indices=True)
selected_feature_names = x.columns[selected_features]
selected_feature_names
```

MODEL CREATION

```python
#splitting datset for training and testing the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#creating model using multiple linear regression to predict the carbon footprint value
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
ypred=model.predict(x_test)
ypred
```

```python
print(list(zip(x,model.coef_)))
```

```python
print("constant is",model.intercept_)
```

```python
df3=pd.DataFrame({'actual value': y_test,'pred value': ypred,'error': y_test-ypred})
df3
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("mae is",mean_absolute_error(y_test,ypred))
print("mape is",mean_absolute_percentage_error(y_test,ypred))
print("r2 score",r2_score(y_test,ypred))
print("mse is",mean_squared_error(y_test,ypred))
```

```python
#creating model using Randomforest regressor
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
regressor.fit(x_train,y_train)
ypred=regressor.predict(x_test)
ypred
```

```python
# Evaluating the model
print("mae is",mean_absolute_error(y_test,ypred))
print("mape is",mean_absolute_percentage_error(y_test,ypred))
print("r2 score",r2_score(y_test,ypred))
print("mse is",mean_squared_error(y_test,ypred))
```

```python
#hyper parameter tuning
from sklearn.model_selection import GridSearchCV

param={'n_estimators':[10,100],'criterion':['squared_error','absolute_error','friedman_mse','poisson']}

gd=GridSearchCV(regressor,param,cv=10,scoring='r2')
gd.fit(x_train,y_train)

```
